Segmenting and Clustering Neighborhoods in Toronto
week3

scrape the following Wikipedia page
(https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M)

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

wiki = requests.get("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
wiki1 = BeautifulSoup(wiki.content, 'html.parser')
```

create the dataframe from wikipedia page

```python
body = wiki1.find('tbody')
rows = body.select('tr')
row = [r.get_text() for r in rows]
df = pd.DataFrame(row)
df1 = df[0].str.split('\n', expand=True)
df1.head(11)
```

```python
df2 = df1.rename(columns=df1.iloc[0])
df3 = df2.drop(df2.index[0])
df3.head(11)
```

ignore cells with a borough that is Not assigned

```python
df4 = df3[df3.Borough != 'Not assigned']
df4.head(11)
```

combine neighborhoods

```python
df5 = df4.groupby(['Postcode', 'Borough'], sort = False).agg(','.join)
df5.reset_index(inplace = True)
df5.head(11)
```

change the value of the neighborhood

```python
df6 = df5.replace("Not assigned", "Queen's Park")
df6.head()
```
